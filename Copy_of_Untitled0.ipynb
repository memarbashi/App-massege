```python
import os
import speech_recognition as sr
from pydub import AudioSegment
from IPython.display import display, Audio
import csv
import subprocess

def text_to_speech(text, lang='fa', filename='output.wav'):
    command = f'espeak -v {lang} -s 150 "{text}" --stdout > {filename}'
    try:
        subprocess.run(command, shell=True, check=True)
        print(f"فایل صوتی ساخته شد: {filename}")
    except subprocess.CalledProcessError as e:
        print(f"خطا در ایجاد فایل صوتی: {e}")

def speech_to_text(audio_file):
    recognizer = sr.Recognizer()
    audio_file_path = os.path.join(os.getcwd(), audio_file)  # افزودن مسیر فایل صریح برای رفع خطا
    if not os.path.exists(audio_file_path):
        print(f"فایل {audio_file_path} یافت نشد.")
        return ""

    audio = AudioSegment.from_wav(audio_file_path)
    audio = audio + 10  # افزایش حجم صدا
    audio.export("temp.wav", format="wav")
    with sr.AudioFile("temp.wav") as source:
        audio_data = recognizer.record(source)
        try:
            text = recognizer.recognize_google(audio_data, language="fa-IR")
        except sr.UnknownValueError:
            print("Could not understand audio")
            return ""
        except sr.RequestError as e:
            print(f"Could not request results from Google Speech Recognition service; {e}")
            return ""
        return text

def get_caller_name(phone_number):
    try:
        with open('contacts.csv', mode='r') as file:
            csv_reader = csv.reader(file)
            for row in csv_reader:
                if row[1] == phone_number:
                    return row[0]
    except FileNotFoundError:
        print("File not found: contacts.csv")
    return "Unknown"

def read_messages_from_file(file_path):
    try:
        with open(file_path, 'r') as file:
            messages = file.readlines()
        return [message.strip() for message in messages]
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return []

if __name__ == "__main__":
    phone_number = "1234567890"
    caller_name = get_caller_name(phone_number)
    text_to_speech(f"تماس از {caller_name}", filename="caller_name.wav")

    messages = read_messages_from_file('messages.txt')
    if not messages:
        print("پیامی برای تبدیل موجود نیست.")
    else:
        for i, message in enumerate(messages):
            text_to_speech(message, filename=f"output_{i+1}.wav")

        print("پخش فایل caller_name.wav:")
        display(Audio("caller_name.wav"))

        for i in range(len(messages)):
            print(f"پخش فایل output_{i+1}.wav:")
            display(Audio(f"output_{i+1}.wav"))

        audio_file = "output_1.wav"
        text = speech_to_text(audio_file)
        print(f"متن تبدیل شده: {text}")
```

```python
import sys
from PyQt5.QtWidgets import QApplication, QMainWindow, QLabel, QVBoxLayout, QWidget, QLineEdit, QPushButton, QTextEdit
import sqlite3
import subprocess
from multiprocessing import Process

# تابع دریافت نام تماس گیرنده از دیتابیس
def get_caller_name(phone_number):
    try:
        conn = sqlite3.connect('database.db')
        cursor = conn.cursor()
        cursor.execute("SELECT name FROM contacts WHERE phone=?", (phone_number,))
        result = cursor.fetchone()
        conn.close()
        if result:
            return result[0]
        return "Unknown"
    except sqlite3.Error as e:
        print(f"Database error: {e}")
        return "Unknown"

# تابع خواندن پیام‌ها از دیتابیس
def read_messages_from_db():
    try:
        conn = sqlite3.connect('database.db')
        cursor = conn.cursor()
        cursor.execute("SELECT message FROM messages")
        messages = cursor.fetchall()
        conn.close()
        return [message[0] for message in messages]
    except sqlite3.Error as e:
        print(f"Database error: {e}")
        return []

class App(QMainWindow):
    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):
        self.setWindowTitle('رابط کاربری گرافیکی')
        self.setGeometry(100, 100, 600, 400)

        central_widget = QWidget(self)
        self.setCentralWidget(central_widget)

        layout = QVBoxLayout(central_widget)

        self.phone_input = QLineEdit(self, placeholderText='شماره تماس')
        self.name_display = QLabel('نام تماس گیرنده: ')
        self.fetch_button = QPushButton('دریافت نام')
        self.message_display = QTextEdit(self)
        self.message_display.setReadOnly(True)

        layout.addWidget(self.phone_input)
        layout.addWidget(self.name_display)
        layout.addWidget(self.fetch_button)
        layout.addWidget(self.message_display)

        self.fetch_button.clicked.connect(self.fetch_caller_name)

    def fetch_caller_name(self):
        phone_number = self.phone_input.text()
        print(f"Fetching caller name for phone number: {phone_number}")  # پیام دیباگ
        caller_name = get_caller_name(phone_number)
        self.name_display.setText(f'نام تماس گیرنده: {caller_name}')
        self.display_messages()

    def display_messages(self):
        messages = read_messages_from_db()
        print(f"Displaying {len(messages)} messages")  # پیام دیباگ
        message_text = "\n".join([f'پیام {i+1}: {message}' for i, message in enumerate(messages)])
        self.message_display.setText(message_text)

def run_streamlit():
    print("Starting Streamlit")  # پیام دیباگ
    subprocess.run(['streamlit', 'run', 'streamlit_app.py'])

if __name__ == '__main__':
    try:
        # اجرای فرآیند PyQt5
        print("Starting PyQt5")  # پیام دیباگ
        pyqt_process = Process(target=lambda: (QApplication(sys.argv).exec_()))
        pyqt_process.start()

        # اجرای فرآیند Streamlit
        print("Starting Streamlit Process")  # پیام دیباگ
        streamlit_process = Process(target=run_streamlit)
        streamlit_process.start()

        # منتظر ماندن برای اتمام هر دو فرآیند
        pyqt_process.join()
        streamlit_process.join()
    except Exception as e:
        print(f"Error occurred: {e}")
```

```python
import csv
import sqlite3

def sync_contacts(csv_file):
    conn = sqlite3.connect('database.db')
    cursor = conn.cursor()

    # ایجاد جدول مخاطبان اگر وجود ندارد
    cursor.execute('''CREATE TABLE IF NOT EXISTS contacts (
                        id INTEGER PRIMARY KEY AUTOINCREMENT,
                        name TEXT NOT NULL,
                        phone TEXT NOT NULL)''')

    with open(csv_file, 'r') as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            name, phone = row
            cursor.execute('''INSERT OR IGNORE INTO contacts (name, phone)
                              VALUES (?, ?)''', (name, phone))

    conn.commit()
    conn.close()
    print("دفترچه مخاطبان همگام‌سازی شد.")
```

```python
from twilio.rest import Client
import vosk
import sounddevice as sd
import sqlite3
import ipywidgets as widgets
from IPython.display import display, HTML, Audio
import json
import queue

q = queue.Queue()

# اطلاعات حساب کاربری Twilio خود را جایگزین کنید
account_sid = 'YOUR_TWILIO_ACCOUNT_SID'
auth_token = 'YOUR_TWILIO_AUTH_TOKEN'
twilio_phone_number = 'YOUR_TWILIO_PHONE_NUMBER'
client = Client(account_sid, auth_token)

def send_sms(to, body):
    """ارسال پیامک با استفاده از Twilio"""
    message = client.messages.create(
        body=body,
        from_=twilio_phone_number,
        to=to
    )
    return message.sid

def get_caller_name(phone_number):
    """دریافت نام تماس‌گیرنده از پایگاه داده"""
    conn = sqlite3.connect('database.db')
    cursor = conn.cursor()
    cursor.execute("SELECT name FROM contacts WHERE phone=?", (phone_number,))
    result = cursor.fetchone()
    conn.close()
    if result:
        return result[0]
    return "نام یافت نشد"

def make_call(to, text):
    """برقراری تماس و پخش پیام صوتی"""
    call = client.calls.create(
        twiml=f'<Response><Say>{text}</Say></Response>',
        from_=twilio_phone_number,
        to=to
    )
    return call.sid

def callback(indata, frames, time, status):
    if status:
        print(status)
    q.put(bytes(indata))

def recognize_speech():
    """تبدیل صدای کاربر به متن با استفاده از Vosk"""
    model = vosk.Model("path_to_your_vosk_model")  # مسیر صحیح مدل Vosk را وارد کنید
    recognizer = vosk.KaldiRecognizer(model, 16000)

    with sd.RawInputStream(samplerate=16000, blocksize=8000, dtype='int16', channels=1, callback=callback):
        print("در حال گوش دادن...")
        while True:
            data = q.get()
            if recognizer.AcceptWaveform(data):
                result = recognizer.Result()
                result = json.loads(result)
                return result.get('text', '')

def read_sms():
    """دریافت پیامک‌های دریافتی و خواندن آن‌ها به صورت صوتی"""
    messages = client.messages.list(to=twilio_phone_number)
    for message in messages:
        print(f"پیام دریافت شده از {message.from_}: {message.body}")
        text_to_speech(message.body, filename=f"msg_{message.sid}.wav")
        display(Audio(f"msg_{message.sid}.wav"))

# ایجاد رابط کاربری با استفاده از ipywidgets
tab = widgets.Tab()
tab.children = [widgets.Output(), widgets.Output(), widgets.Output(), widgets.Output(), widgets.Output()]
tab.set_title(0, 'جستجوی تماس‌گیرنده')
tab.set_title(1, 'ارسال پیامک')
tab.set_title(2, 'برقراری تماس')
tab.set_title(3, 'خواندن پیامک‌ها')
tab.set_title(4, 'ارسال پیامک با صدا')

# تب جستجوی تماس‌گیرنده
with tab.children[0]:
    phone_input = widgets.Text(description='شماره تلفن:', layout=widgets.Layout(width='100%'))
    display(phone_input)

    button = widgets.Button(description='جستجوی نام', layout=widgets.Layout(width='100%'))
    caller_name_output = widgets.Output()

    def on_button_click(b):
        with caller_name_output:
            caller_name_output.clear_output()
            caller_name = get_caller_name(phone_input.value)
            print(f"نام تماس‌گیرنده: {caller_name}")

    button.on_click(on_button_click)
    display(button, caller_name_output)

# تب ارسال پیامک
with tab.children[1]:
    phone_input_for_sms = widgets.Text(description='شماره گیرنده:', layout=widgets.Layout(width='100%'))
    display(phone_input_for_sms)

    message_input = widgets.Text(description='پیام:', layout=widgets.Layout(width='100%'))
    display(message_input)

    send_button = widgets.Button(description='ارسال پیام', layout=widgets.Layout(width='100%'))
    send_message_output = widgets.Output()

    def send_sms_button_clicked(b):
        with send_message_output:
            send_message_output.clear_output()
            to = phone_input_for_sms.value
            body = message_input.value
            try:
                message_sid = send_sms(to, body)
                print(f"پیام با موفقیت ارسال شد. شناسه پیام: {message_sid}")
            except Exception as e:
                print(f"خطا در ارسال پیام: {str(e)}")

    send_button.on_click(send_sms_button_clicked)
    display(send_button, send_message_output)

# تب برقراری تماس
with tab.children[2]:
    phone_input_for_call = widgets.Text(description='شماره گیرنده:', layout=widgets.Layout(width='100%'))
    display(phone_input_for_call)

    call_message_input = widgets.Text(description='پیام صوتی:', layout=widgets.Layout(width='100%'))
    display(call_message_input)

    call_button = widgets.Button(description='برقراری تماس', layout=widgets.Layout(width='100%'))
    call_output = widgets.Output()

    def make_call_button_clicked(b):
        with call_output:
            call_output.clear_output()
            to = phone_input_for_call.value
            text = call_message_input.value
            try:
                call_sid = make_call(to, text)
                print(f"تماس برقرار شد. شناسه تماس: {call_sid}")
            except Exception as e:
                print(f"خطا در برقراری تماس: {str(e)}")

    call_button.on_click(make_call_button_clicked)
    display(call_button, call_output)

# تب خواندن پیامک‌ها
with tab.children[3]:
    read_sms_button = widgets.Button(description='خواندن پیامک‌ها', layout=widgets.Layout(width='100%'))
    read_sms_output = widgets.Output()

    def read_sms_button_clicked(b):
        with read_sms_output:
            read_sms_output.clear_output()
            try:
                read_sms()
            except Exception as e:
                print(f"خطا در خواندن پیامک‌ها: {str(e)}")

    read_sms_button.on_click(read_sms_button_clicked)
    display(read_sms_button, read_sms_output)

# تب ارسال پیامک با صدا
with tab.children[4]:
    voice_message_output = widgets.Output()
    voice_message_button = widgets.Button(description='تشخیص صدا', layout=widgets.Layout(width='100%'))

    def voice_message_button_clicked(b):
        with voice_message_output:
            voice_message_output.clear_output()
            text = recognize_speech()
            if text:
                message_input.value = text

    voice_message_button.on_click(voice_message_button_clicked)
    display(voice_message_button, voice_message_output)

display(tab)
```
